A notebook to analyze my Neflix account with data cleaning and visulization.

Loading necessary libraries for the analysis.

```python
#libraries for anyalsis

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime,date



```

Importing the first csv file using the Pandas library while using the read function within the library. Setting the dataframe index to the Title column of the file.

```python
#Reading in the first csv file and setting the index for the dataframe

df = pd.read_csv('ViewingActivity.csv')
df = df.set_index('Title')
df[:30]
```

Checking the dataframe shape.

```python
df.shape
```

Listing the column headers to see what needs to be removed for the analysis.

```python
df.columns
```

Creating a new dataframe by dropping unwanted coulumns and setting the display with to none. Also setting the Duration column to read time Delta for easier translation of time. Setting display width.

```python
#Dropping columns and setting time to new format

newdf = df.drop(['Attributes','Bookmark','Latest Bookmark'], axis = 1)
pd.set_option('display.width',None)
newdf['Duration'] = pd.to_timedelta(newdf['Duration'])
newdf
```

Checking to see if any Na values exist in the dataframe.

```python
# Seeing if any NaN values are in the dataframe

newdf[newdf.isna().any(axis=1)]
```

Replacing the all NaN values for Device Type with Unknown and Supplemental Video Type  with Orignal in the columns.

```python
# Changing NaN values to different values.

newdf['Device Type'].replace(np.nan,'Unknown',inplace =True)
newdf['Supplemental Video Type'].replace(np.nan,'Original',inplace =True)


newdf
```

Renaming Start Time column to Date.

```python
#Renaming the start time column to Date

newdf.rename(columns ={'Start Time': 'Date'}, inplace= True)
newdf
```

Changing the Date column format to read just the date eliminating the time stamp.

```python
#Eliminaing the time from the Date column

newdf['Date'] = pd.to_datetime(newdf['Date']).dt.normalize()
newdf
```

Checking the dataframe for any other NaN values.

```python
# Checking to see if all NaN values are resolved in the dataframe

newdf[newdf.isna().any(axis=1)]
```

Checking dataframe for any duplicates.

```python
newdf.duplicated()
```

Dropping any duplicate rows for dataframe.

```python
newdf.drop_duplicates(inplace=True)
newdf
```

Eliminating any rows with the word TRAILER,HOOK and Promotional listed in the Supplemental Video Column. They are short clips of shows or movies not needed for analysis.

```python
newdf = newdf[newdf['Supplemental Video Type'].str.contains('TRAILER') ==False]
newdf = newdf[newdf['Supplemental Video Type'].str.contains('HOOK') ==False]
newdf = newdf[newdf['Supplemental Video Type'].str.contains('Promotional') ==False]
newdf
```

Grouping profile names to get a count of titles veiwed of the dataset.

```python
users = newdf.groupby('Profile Name', as_index=True).size()
users
```


By using the Matplotlib library I plotted the profile names to show visulization of how many titles each profile has watched over the years. 

```python
plt.figure(figsize= (20,10))
plt.title('User Profiles that are in the household', fontsize = 30)
plt.ylabel('Titles Viewed', fontsize = 30)
plt.xlabel('Users', fontsize=35)
users.plot(kind='bar',color="red") 
```

Checking what the count is for all my titles in my dataframe.

```python
#Counts of titles viewed in dataset
titles_viewed= newdf.index.value_counts()
titles_viewed[1:11]
```

Using the seaborn library I plotted the top 10 most titles watched.  

```python
# Plotting the top 10 titles watched by viewers using the seaborn library

titles_viewed= newdf.index.value_counts()
titles_viewed = titles_viewed[1:11]

plt.figure(figsize = (10,15))
sns.set_style("whitegrid")
ax = sns.barplot(y= titles_viewed.index, x=titles_viewed, palette = "Oranges_r", orient= 'h')
ax.set_ylabel('Titles',size = 18)
ax.set_xlabel("Most Viewed", size = 18)
ax.set_title('Most Titles Viewed by Users',fontsize = 22)
```

Importing the second csv file using the Pandas library while using the read function within the library

```python
# Importing the second csv file for anlysis

titles = pd.read_csv('titles.csv')

titles
```

Examining the columns to see what is needed.

```python
titles.columns
```

Dropping unnecessary columns from dataframe.

```python
titles = titles.drop(['description','imdb_id','imdb_score','imdb_votes','tmdb_popularity','tmdb_score'], axis = 1)
titles
```

Checking dataframe for any NaN values.

```python
titles = titles[titles.isna().any(axis=1)]
titles
```

Replacing the seasons column NaNs with 0. Also put in the pandas mode option to get rid of warnings for the dataframe.

```python
pd.options.mode.chained_assignment = None
titles['seasons'] =titles['seasons'].replace(np.nan, 0)
titles
```

Replacing the age certification colum NaNs with NR.

```python
titles['age_certification'].replace(np.nan,'NR',inplace =True)
titles
```

Changing lower case title column to Title and changing production_countries to just Country.

```python
titles.rename(columns ={'title': 'Title','production_countries': 'Country'}, inplace= True)
titles= titles.set_index('Title')
titles
```

Created new dataframe by merging the first and second dataframes which resulted in dropping all the tv shows in the data and just matching the movies that are in the first dataset while adding the release year among other information.

```python

mergedtitles = pd.merge (newdf, titles, left_index=True, right_index=True)
mergedtitles

```

Dropping any remaining duplicates in the new dataframe.  It appears that only one was a duplicate.

```python
mergedtitles.drop_duplicates(inplace=True)
mergedtitles
```

Grouped age certification column to see how many movie titles watched by rating.

```python
# Grouping the age_certification column 
mergedtitles.groupby('age_certification', as_index=True).size()
```

Creating a dataframe for movie ratings for movies watched.


```python
# Another way to group age certification column but in order of most to least rating for movies wathced
movie_ratings = mergedtitles['age_certification'].value_counts()
movie_ratings
```
