Loading necessary libraries for the analysis.

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import numpy as np
from datetime import datetime,date



```

Importing the first csv file using the Pandas library while using the read function within the library. Setting the dataframe index to the Title column of the file.

```python
df = pd.read_csv('ViewingActivity.csv')
df = df.set_index('Title')
df[:30]
```

Checking the dataframe shape.

```python
df.shape
```

Listing the column headers to see what needs to be removed for the analysis.

```python
df.columns
```

Creating a new dataframe by dropping unwanted coulumns and setting the display with to none.

```python
newdf = df.drop(['Attributes','Bookmark','Latest Bookmark'], axis = 1)
pd.set_option('display.width',None)
newdf
```

Checking to see if any Na values exist in the dataframe.

```python
newdf[newdf.isna().any(axis=1)]
```

Replacing the all NaN values for Device Type with Unknown and Supplemental Video Type  with Orignal in the columns.

```python
newdf['Device Type'].replace(np.nan,'Unknown',inplace =True)
newdf['Supplemental Video Type'].replace(np.nan,'Original',inplace =True)


newdf
```

Renaming Start Time columns to Date.

```python
newdf.rename(columns ={'Start Time': 'Date'}, inplace= True)
newdf
```

Changing the Date column format to read just the date eliminating the time stamp.

```python
newdf['Date'] = pd.to_datetime(newdf['Date']).dt.normalize()
newdf
```

Checking the dataframe for any other NaN values.

```python
newdf[newdf.isna().any(axis=1)]
```

Checking dataframe for any duplicates.

```python
newdf.duplicated()
```

Dropping any duplicate rows for dataframe.

```python
newdf.drop_duplicates(inplace=True)
newdf
```

Eliminating any rows with the word TRAILER,HOOK and Promotional listed in the Supplemental Video Column. They are short clips of shows or movies not needed for analysis.

```python
newdf = newdf[newdf['Supplemental Video Type'].str.contains('TRAILER') ==False]
newdf = newdf[newdf['Supplemental Video Type'].str.contains('HOOK') ==False]
newdf = newdf[newdf['Supplemental Video Type'].str.contains('Promotional') ==False]
newdf
```

Checking device counts to see wich has been used most.

```python
newdf['Device Type'].value_counts()
```

Importing the second csv file using the Pandas library while using the read function within the library

```python
titles = pd.read_csv('titles.csv')

titles
```

Examining the columns to see what is needed.

```python
titles.columns
```

Dropping unnecessary columns from dataframe.

```python
titles = titles.drop(['description','imdb_id','imdb_score','imdb_votes','tmdb_popularity','tmdb_score'], axis = 1)
titles
```

Checking dataframe for any NaN values.

```python
titles = titles[titles.isna().any(axis=1)]
titles
```

Replacing the seasons column NaNs with 0. Also put in the pandas mode option to get rid of warnings for the dataframe.

```python
pd.options.mode.chained_assignment = None
titles['seasons'] =titles['seasons'].replace(np.nan, 0)
titles
```

Replacing the age certification colum NaNs with NR.

```python
titles['age_certification'].replace(np.nan,'NR',inplace =True)
titles
```

Changing lower case title column to Title and changing production_countries to just Country.

```python
titles.rename(columns ={'title': 'Title','production_countries': 'Country'}, inplace= True)
titles= titles.set_index('Title')
titles
```

```python
mergedtitles = pd.merge (newdf, titles, left_index=True, right_index=True)
#mergedtitles = newdf.merge(titles,how ='outer')
mergedtitles[0:100]

```
